EDA Categorical Variables

```python
# Find the number of unique values for each categorical variable
def unique_count(name, df):
  counts = df.nunique()
  return f"{name}: {counts}"
```

```python
print(unique_count("ISBN", books['ISBN']))
print(unique_count("Book-Title", books['Book-Title'])) # important to note that there are book titles with the same name for different books, we should tell different books apart by ISBN number.
print(unique_count("Book-Author", books['Book-Author']))
print(unique_count("Year-Of-Publication", books['Year-Of-Publication']))
print(unique_count("Publisher", books['Publisher']))
print(unique_count("Series", books['Series']))
```

```python
def topNChart(name, df, N, by):
  top10books = df.groupby(name).size().nlargest(N)
  y = top10books.keys()
  x = top10books.to_numpy()

  bars = plt.barh(y, x)
  plt.bar_label(bars, labels=x, padding = 2, color='black', fontsize=8)
  plt.title(f"Top {N} {name} by {by}")
  plt.ylabel(name)
  plt.xlabel('Count')
  plt.show()
```

```python
# Top 10 Book-Author By Book Count
topNChart('Book-Author', books, 10, "Book Count")
```

```python
# Top 10 Publisher By Book Count
topNChart('Publisher', books, 10, "Book Count")
```

```python
# Top 10 Series By Book Count
topNChart('Series', books, 10, "Book Count")
```

```python
ls = []
for col in genres.columns:
  genre_ls = [col, genres[col].sum()]
  ls.append(genre_ls)
genres_count = pd.DataFrame(ls, columns = ['Genres', 'Count'])
genres_count
```

```python
top10genres = genres_count.sort_values(by = 'Count', ascending = False)[:10]
y = top10genres['Genres']
x = top10genres['Count']

bars = plt.barh(y, x)
plt.bar_label(bars, labels=x, padding = 2, color='black', fontsize=8)
plt.title("Top Genres By Book Count")
plt.ylabel('Genres')
plt.xlabel('Count')
plt.show()
```

```python
# Top 10 Countries By Users
topNChart('Country', users, 10, "User Count")
```
